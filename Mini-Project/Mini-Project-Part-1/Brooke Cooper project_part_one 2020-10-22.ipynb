```python
# Dependencies
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# load CSV
file_to_load = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coders_df = pd.read_csv(file_to_load)
```

```python
# Inspect all columns
coders_df.columns
```

```python
# Inspect columns of original data file (Part 2)
# Create a series object containing the data type objects of each column of Dataframe.
coders_type = coders_df.dtypes

print('Number of data types by column of original data set Dataframe :')
coders_type.value_counts()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
coders_x_df = coders_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
coders_x_df.head()
```

```python
# Inspect all columns
coders_x_df.columns
```

```python
# Inspect columns of extraced data
# Create a series object containing the data type objects of each column of Dataframe.
coders_x_type = coders_x_df.dtypes

print('Number of data types by column of extracted data Dataframe :')
coders_x_type.value_counts()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
coders_x_df["AttendedBootcamp"] = coders_x_df["AttendedBootcamp"].replace({0: "No", 1 : "Yes"})
coders_x_df.head()
```

```python
# Calculate total number of respondents in survey
total_respondents = coders_x_df["AttendedBootcamp"].count()
total_respondents
```

```python
# Calculate total number of respondents who attended a bootcamp
print("Respondents who Attended Bootcamp = 'Yes':")
coders_x_df["AttendedBootcamp"].value_counts()
```

```python
# Extract rows corresponding only to people who attended a bootcamp
coders_x_attend_df = coders_x_df.loc[coders_x_df["AttendedBootcamp"] == "Yes"]
coders_x_attend_df.head()
```

```python
# Use describe to profile the attendee DataFrame
coders_x_attend_df.describe()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
coders_x_attend_df["BootcampFinish"] = coders_x_attend_df["BootcampFinish"].replace({0: "No", 1 : "Yes"})
coders_x_attend_df.head()
```

```python
# Calculate total number of respondents who both attended and completed a bootcamp
coders_x_finish = coders_x_attend_df["BootcampFinish"].value_counts()
coders_x_finish_df = pd.DataFrame(coders_x_finish)
print("Respondents who Attended and Completed Bootcamp = 'Yes':")
coders_x_finish_df
```

```python
# Single field referring to total number of respondents who both attended and completed a bootcamp
# Single field referring to total number of respondents who answered the completion survey question
coders_finish = coders_x_finish_df.iloc[0,0]
finish_respondents = coders_x_finish_df.iloc[0,0] + coders_x_finish_df.iloc[1,0]
coders_finish_pct = coders_x_finish_df.iloc[0,0] / finish_respondents
print(coders_finish)
print(coders_finish_pct)
print(finish_respondents)
```

```python
# Calculate average age of attendees
boot_attend_avg_age = coders_x_attend_df["Age"].mean()

# Calculate how many people attended a bootcamp
boot_attend_ttl = coders_x_attend_df["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees
boot_attend_degrees = coders_x_attend_df["SchoolDegree"].value_counts()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
boot_attend_gender = coders_x_attend_df["Gender"].value_counts()

# Calculate percentage of respondents who attended a bootcamp
percent_attended = boot_attend_ttl/total_respondents

# Calculate percentage of respondents belonging to each gender
boot_attend_gender_pct = coders_x_attend_df["Gender"].value_counts(normalize=True).mul(100).round(1).astype(str) + "%"

# Calculate percentage of attendees with a school degree
boot_attend_degrees_pct = coders_x_attend_df["SchoolDegree"].value_counts(normalize=True).mul(100).round(1).astype(str) + "%"

# Calculate average post-bootcamp salary
boot_attend_avg_salary_post = coders_x_attend_df["BootcampPostSalary"].mean()

# Print results of above calculations
print(f"Average Age = {boot_attend_avg_age}\n")
print(f"Number of People Who Attended Bootcamp = {boot_attend_ttl}\n")
print(f"School Degree Profile of Bootcamp Attendees: \n")
print(f"{boot_attend_degrees}\n")
print("Gender Profile of Bootcamp Attendees: \n")
print(f"{boot_attend_gender}\n")
print(f"Attendees Percentage of Total Survey Respondents = {percent_attended})\n")
print("Gender Profile as a Percentage of Total Bootcamp Attendees: \n")
print(f"{boot_attend_gender_pct}\n")
print("School Degree Profile as a Percentage of Total Bootcamp Attendees: \n")
print(f"{boot_attend_degrees_pct}\n")
print(f"Average Salary Post-Bootcamp = {boot_attend_avg_salary_post}\n")
```

```python
# Create a new table consolodating above calculations

#Convert value_counts() to DataFrames
boot_attend_degrees_df = pd.DataFrame(boot_attend_degrees)
boot_attend_degrees_cvt1_df = boot_attend_degrees_df.reset_index()
boot_attend_degrees_cvt2_df = boot_attend_degrees_cvt1_df.rename(columns={"SchoolDegree": "Number"})
boot_attend_degrees_cvt2_df = boot_attend_degrees_cvt2_df.rename(columns={"index": "School Degree"})

boot_attend_degrees_pct_df = pd.DataFrame(boot_attend_degrees_pct)
boot_attend_degrees_pct_cvt1_df = boot_attend_degrees_pct_df.reset_index()
boot_attend_degrees_pct_cvt2_df = boot_attend_degrees_pct_cvt1_df.rename(columns={"SchoolDegree": "Percentage"})
boot_attend_degrees_pct_cvt2_df = boot_attend_degrees_pct_cvt2_df.rename(columns={"index": "School Degree"})

boot_attend_gender_df = pd.DataFrame(boot_attend_gender)
boot_attend_gender_cvt1_df = boot_attend_gender_df.reset_index()
boot_attend_gender_cvt2_df = boot_attend_gender_cvt1_df.rename(columns={"Gender": "Number"})
boot_attend_gender_cvt2_df = boot_attend_gender_cvt2_df.rename(columns={"index": "Gender"})

boot_attend_gender_pct_df = pd.DataFrame(boot_attend_gender_pct)
boot_attend_gender_pct_cvt1_df = boot_attend_gender_pct_df.reset_index()
boot_attend_gender_pct_cvt2_df = boot_attend_gender_pct_cvt1_df.rename(columns={"Gender": "Percentage"})
boot_attend_gender_pct_cvt2_df = boot_attend_gender_pct_cvt2_df.rename(columns={"index": "Gender"})
```

```python
degree_dict = {"School Degree": ["bachelor's degree",
               "some college credit, no degree",
               "master's degree (non-professional)",
               "professional degree (MBA, MD, JD, etc.)",
               "high school diploma or equivalent (GED)",
               "associate's degree",
               "trade, technical, or vocational training",
               "some high school",
               "Ph.D.",
               "no high school (secondary school)"],
               "Degree": ["Yes", "No", "Yes", "Yes", "No", "No", "No", "No", "Yes", "No"]}
    

# {"bacherlor's degree": "Yes, 
#               "some college credit, no degree": "No", 
#               "master's degree (non-professional)": "Yes",
#               "professional degree (MBA, MD, JD, etc.)": "Yes",
#               "high school diploma or equivalent (GED)": "No",
#               "associate's degree": "No",
#               "trade, technical, or vocational training": "No",
#               "some high school": "No",
#               "Ph.D.": "Yes",
#               "no high school (secondary school)": "No"}

degree_df = pd.DataFrame.from_dict(degree_dict)
degree_df
```

```python
boot_attend_gender_cvt2_df
```

```python
boot_attend_gender_pct_cvt2_df
```

```python
# Merge two dataframes using an inner join - Degree DataFrames (on = School Degree)
data_frames1 = [boot_attend_degrees_cvt2_df, boot_attend_degrees_pct_cvt2_df, degree_df]
boot_attend_degree_merge_df = reduce(lambda  left,right: pd.merge(left,right,on=["School Degree"], how="outer"), data_frames1)
boot_attend_degree_merge_df
```

```python
# Create a GroupBy object based upon "Degree"
group_by_degree = boot_attend_degree_merge_df.groupby("Degree")

# Numbers of attendees by degree status 
group_by_degree = group_by_degree["Number"].sum()
group_by_degree_df = pd.DataFrame(group_by_degree)
group_by_degree_df
```

```python
# Extract number of fully degreed bootcampers (not including associate degrees)
full_degree_yes_number = group_by_degree_df.iloc[1, 0]
full_degree_yes_number
```

```python
# Calculate percentage of degreed bootcampers (not including associate degrees)
total_degree_respondents = full_degree_yes_number + group_by_degree_df.iloc[0, 0]
full_degree_yes_pct = full_degree_yes_number / total_degree_respondents

full_degree_yes_pct
```

```python
assoc_degree_number = boot_attend_degree_merge_df.iloc[5, 1]
assoc_degree_number
```

```python
# Calculate percentage of degreed bootcampers including associate degrees
degree_includ_assoc_pct = (full_degree_yes_number + assoc_degree_number) / total_degree_respondents
degree_includ_assoc_pct
```

```python
# Merge two dataframes using an inner join - Gender DataFrames (on = Gender)
boot_attend_gender_merge_df = pd.merge(boot_attend_gender_cvt2_df, boot_attend_gender_pct_cvt2_df, on="Gender")
boot_attend_gender_merge_df
```

```python
boot_attend_gender_respond = boot_attend_gender_merge_df["Number"].sum()
boot_attend_gender_respond
```

```python
# Extract percentage of male bootcamper respondents
male_boot_attend_pct = boot_attend_gender_merge_df.iloc[0, 1] / boot_attend_gender_respond
male_boot_attend_pct
```

```python
# Extract percentage of female bootcamper respondents
female_boot_attend_pct = boot_attend_gender_merge_df.iloc[1, 1] / boot_attend_gender_respond
female_boot_attend_pct
```

```python
# Calculate percentage of Other or Not-Specified Gender bootcamper respondents
other_gender_not_spec_boot_attend_pct = \
    1 - ((boot_attend_gender_merge_df.iloc[0, 1] + boot_attend_gender_merge_df.iloc[1, 1]) / boot_attend_gender_respond)
other_gender_not_spec_boot_attend_pct
```

```python
# FINALLY!!!  A new table consolodating above calculations -- without formatting
# Create a DataFrame of elements analyzed using a list of dictionaries, in this case a single dictionary
boot_analysis_list = [
    {"Average Age": boot_attend_avg_age, 
     "Bootcampers": boot_attend_ttl, 
     "Bootcamper % Respondents": percent_attended * 100, 
     "Avg Salary Post-BC": boot_attend_avg_salary_post,
     "Fully Degreed BCers %": full_degree_yes_pct * 100,
     "Male BCer %": male_boot_attend_pct * 100,
     "Female BCer %": female_boot_attend_pct * 100,
     "Other Gen / Not Spec %": other_gender_not_spec_boot_attend_pct * 100}]
boot_analysis_df = pd.DataFrame(boot_analysis_list)

boot_analysis_df
```

```python
# Improve formatting before outputting spreadsheet
boot_analysis_sty_df = boot_analysis_df.style.format({"Average Age": "{:,.1f}", 
                                                      "Bootcamper % Respondents": "{:,.1f}%", 
                                                      "Avg Salary Post-BC": "${:,.0f}", 
                                                      "Fully Degreed BCers %": "{:,.1f}%",
                                                      "Male BCer %": "{:,.1f}%",
                                                      "Female BCer %": "{:,.1f}%",
                                                      "Other Gen / Not Spec %": "{:,.1f}%"})

boot_analysis_sty_df.hide_index()
```

```python
# Assemble bootcamper respondent counts by specific question that were not created and named earlier
boot_salary_respondents = coders_x_attend_df["BootcampPostSalary"].count()
boot_age_respondents = coders_x_attend_df["Age"].count()
boot_gender_respondents = coders_x_attend_df["Gender"].count()

#Print a highlights of the metadata analysis specific to the 
print(f"Metadata Analysis Highlights")
print(f"----------------------------------------------------------------------------------------\n")
print(f"There were {'{:,.0f}'.format(total_respondents)} total respondents to the survey, of whom {boot_attend_ttl} or {'{:,.1f}'.format(percent_attended * 100)}% attended bootcamp.\n")
print(f"Of the {boot_attend_ttl} who attended bootcamp:\n")
print(f"     {coders_finish} or {'{:,.1f}'.format(coders_finish_pct * 100)}% completed bootcamp of the {finish_respondents} or {'{:,.1f}'.format(finish_respondents / boot_attend_ttl * 100)}% who answered the question")
print(f"     {total_degree_respondents} or {'{:,.1f}'.format(total_degree_respondents / boot_attend_ttl * 100)}% answered the question on education level")
print(f"     {boot_salary_respondents} or {'{:,.1f}'.format(boot_salary_respondents / boot_attend_ttl * 100)}% answered the question on post-bootcamp salary")
print(f"     {boot_age_respondents} or {'{:,.1f}'.format(boot_age_respondents / boot_attend_ttl * 100)}% answered the question on age")
print(f"     {boot_gender_respondents} or {'{:,.1f}'.format(boot_gender_respondents / boot_attend_ttl * 100)}% answered the question on gender")
```
